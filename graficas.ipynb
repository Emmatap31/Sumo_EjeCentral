```python
import xml.etree.ElementTree as ET
import matplotlib.pyplot as plt
```

```python
# Cargar el archivo tripinfo.xml
tree = ET.parse("tripinfo1.xml")
root = tree.getroot()

tiempos_viaje = []

# Extraer el tiempo de cada auto
```

```python
for trip in root.findall("tripinfo"):
    duracion = float(trip.get("duration"))
    tiempos_viaje.append(duracion)

```

```python
# Graficar histograma de tiempos de viaje
plt.hist(tiempos_viaje, bins=20, edgecolor="black")
plt.xlabel("Tiempo en la simulación (segundos)")
plt.ylabel("Número de autos")
plt.title("Distribución de tiempos de viaje")
plt.grid()
plt.show()
```

```python
import pandas as pd

df = pd.DataFrame({"Tiempo de viaje (s)": tiempos_viaje})
df.to_csv("tiempos_viaje.csv", index=False)
```
